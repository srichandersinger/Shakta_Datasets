```python
import os
import re
import pandas as pd
from indic_transliteration import sanscript
from indic_transliteration.sanscript import transliterate
```

```python
def Durga_Sapta_List():
    liner=[]
    final=[]
    for i in range(1,14):
        filename= 'Ch'+str(i)+'_Dev.txt'
        with open(filename, encoding="utf8") as file_object:
            for line in file_object:
                liner.append(line)
    liner = list(map(lambda s: s.strip(), liner))
    liner = list(filter(lambda x: x != '', liner))
    for i in range(0, len(liner)):
        split_string = re.split(r' ॥ |॥', liner[i])
        verse = split_string[0] + '॥'
        verse_number = split_string[1].split('.')
        final.append([transliterate(verse_number[0], sanscript.DEVANAGARI, sanscript.ITRANS), transliterate(verse_number[1], sanscript.DEVANAGARI, sanscript.ITRANS), verse])
    return final
        
```

```python
list1= Durga_Sapta_List()
df = pd.DataFrame(list1, columns=['Chapter', 'Verse Number', 'Verse'])
```

```python
df['Chapter']
```

```python
df.to_csv('Durga_Saptashati.csv', encoding='utf-8', index=False)
```

```python
def is_verse_pattern(s):
    return re.match(r'^\d+(\.\d+(-\d+)?)? ', s.strip())
    
lines=[]
with open('CS_T_Full.txt', encoding='utf-8') as file_object:
    for line in file_object:
        lines.append(line)
lines=[s.strip() for s in lines if is_verse_pattern(s)]        
lines
```

```python
finall=[]
for item in lines:
    # Split the item into the chapter/verse part and the text part
    parts = item.split(' ', 1)
    chapter_verse = parts[0].split('.')
    finall.append([chapter_verse[0], chapter_verse[1], parts[1]])
```

```python
df = pd.DataFrame(finall, columns=['Chapter', 'Verse Number', 'Translation'])
```

```python
df.to_csv('Durga_Saptashati_Translations.csv', encoding='utf-8', index=False)
```
